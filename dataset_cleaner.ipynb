# Data Cleaning Notebook

We choose to make a slope graph of the Male to Female Ratio for selected majors within 4 colleges inside UIUC. We chose this type of graph because it shows the change in gender ratio over a large period of time rather than every year which makes for a less aesthetic visualization (in my opinion)

## Data Cleaning Process
1. Grouped data by college then year. 
2. Filtered data for our selected majors within each college. Our criteria for majors was size and how well known it is (ie everyone knows mathematics is a major but not everyone knows the Agricultural Communications major. For our own ease, we selected the start year as 2004 because many majors were moved into different colleges in the years prior and 2004 is the year that Computer Science was moved from College of Business to College of Engineering. 
3. Fill all N/A values with 0
4. We calculated the M/F Ratio and wrote the DataFrame for each college to its respective csv. We opted for individual files for each college for ease of loading into d3 as pandas is much MUCH easier to use than javascript.

Note: Sorry for the messy code.

```python
import pandas as pd
colleges = ["ACES", "Business", "LAS", "Engineering"]
years = [2004, 2018]

# df = pd.read_csv('uiuc-students-by-curriculum.csv')
# df.sort_values(['Fall', 'College', 'Major Name'], ascending = [True, True, True])
```

```python
pd.options.display.max_rows = 4000
majors = ["Human Development & Family St", "Agr & Consumer Economics", "Animal Sciences", "Crop Sciences", "Nutritional Sciences", 
          "Food Science & Human Nutrition", "Horticulture", "Natural Resrcs & Environ Sci"]
df2 = pd.read_csv('uiuc-students-by-curriculum.csv', thousands=',')
df2 = df2[['Fall','College','Major Name', 'Total','Male', 'Female']]
df2 = df2.loc[df2['College'] == colleges[0]]
df2 = df2.loc[df2['Fall'] == years[0]]
df2 = df2.loc[df2['Major Name'].isin(majors)]
df2["Total"] = df2["Total"].str.replace(",", "").astype(float)
# df2["Male"] = df2["Male"].str.replace(",", "").astype(float)
df2["Female"] = df2["Female"].str.replace(",", "").astype(float) 

df2 = df2.fillna(0)
# df2 = df2[df2.Total > 50]
# df2.sort_values(['Fall','College', 'Major Name'], ascending = [True, True, True])
df2['Total'] = df2.groupby(['Fall', 'Major Name'])['Total'].transform('sum')
#df2['Total'] = df2['new total']
df2['Male'] = df2.groupby(['Fall', 'Major Name'])['Male'].transform('sum')
df2['Female'] = df2.groupby(['Major Name'])['Female'].transform('sum')
#df2['Male'] = df2['male total']
#df2['Female'] = df2['female total']
df2 = df2.drop_duplicates(['Major Name', 'Total', 'Fall'])
df2 = df2.sort_values(['Total'], ascending = [False])
df2['M/F Ratio'] = df2['Male'] / df2['Female']

df = pd.read_csv('uiuc-students-by-curriculum.csv', thousands=',')
df = df[['Fall','College','Major Name', 'Total','Male', 'Female']]
df = df.loc[df['College'] == colleges[0]]
df = df.loc[df['Fall'] == years[1]]
df = df.loc[df['Major Name'].isin(majors)]

df["Total"] = df["Total"].str.replace(",", "").astype(float)
# df2["Male"] = df2["Male"].str.replace(",", "").astype(float)
df["Female"] = df["Female"].str.replace(",", "").astype(float) 

df = df.fillna(0)
# df2 = df2[df2.Total > 50]
# df2.sort_values(['Fall','College', 'Major Name'], ascending = [True, True, True])
df['Total'] = df.groupby(['Fall', 'Major Name'])['Total'].transform('sum')
#df2['Total'] = df2['new total']
df['Male'] = df.groupby(['Fall', 'Major Name'])['Male'].transform('sum')
df['Female'] = df.groupby(['Fall', 'Major Name'])['Female'].transform('sum')
#df2['Male'] = df2['male total']
#df2['Female'] = df2['female total']
df = df.drop_duplicates(['Major Name', 'Total', 'Fall'])
df = df.sort_values(['Total'], ascending = [False])
df['M/F Ratio'] = df['Male'] / df['Female']
df2 = df2.append(df)
df2 = df2.drop(columns = ['College'])
df2['M/F Ratio'] = df2['M/F Ratio'].round(2)
df2.columns = ['Fall', 'MajorName','Total', 'Male', 'Female', 'MFRatio']
df2.to_csv('cleaned_ACES_Majors.csv')
df2
#df2[df2.Total.astype(int) > 30]
#df2.Total.min()
```
