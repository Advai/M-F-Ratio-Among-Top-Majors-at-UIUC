```python
import pandas as pd

pd.options.display.max_rows = 4000

df2 = pd.read_csv('uiuc-students-by-curriculum.csv', thousands=',')
df2 = df2[['Fall','College','Major Name', 'Total','Male', 'Female']]
colleges = ["Engineering"]
years = ['2004']
df2 = df2.loc[df2['College'].isin(colleges)]
df2 = df2.loc[df2['Fall'].isin(years)]
df2["Total"] = df2["Total"].str.replace(",", "").astype(float) 
df2 = df2.fillna(0)
# df2 = df2[df2.Total > 50]
# df2.sort_values(['Fall','College', 'Major Name'], ascending = [True, True, True])
df2['new total'] = df2.groupby(['Fall', 'Major Name'])['Total'].transform('sum')
df2['Total'] = df2['new total']
df2 = df2.drop_duplicates(['Major Name', 'Total', 'Fall'])
# df2.sort_values(['Fall','College', 'Major Name'], ascending = [True, True, True])
df2 = df2.sort_values(['Total'], ascending = [False])

df2
#df2[df2.Total.astype(int) > 30]
#df2.Total.min()
```
